```python
from pyspark import SparkContext
```

```python
sc = SparkContext()
```

```python
rdd = sc.textFile("word.txt")
```

```python
rdd.first()
```

```python
rdd.collect()
```

```python
rdd2 = rdd.flatMap(lambda x: x.split())
```

```python
rdd2.take(4)
```

```python
rdd3 = rdd2.map(lambda x: (x, 1))
```

```python
rdd3.take(4)
```

```python
rdd4 = rdd3.reduceByKey(lambda a, b: a+b)
```

```python
rdd4.collectAsMap()
```

```python
rdd4.saveAsTextFile("wc_result")
```
